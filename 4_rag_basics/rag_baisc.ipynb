```java
import dev.langchain4j.data.document.Document;
import dev.langchain4j.data.document.parser.apache.pdfbox.ApachePdfBoxDocumentParser;
import java.io.FileInputStream;
import java.nio.file.Path;
import java.nio.file.Paths;

try {
    // १. PDF फाईलचा पाथ सेट करा
    Path pdfPath = Paths.get("telecom_guide.pdf");

    // २. डॉक्युमेंट पार्सर सुरू करा
    ApachePdfBoxDocumentParser parser = new ApachePdfBoxDocumentParser();

    // ३. दुरुस्त केलेली ओळ: FileInputStream चा वापर करून PDF पार्स करणे
    Document document = parser.parse(new FileInputStream(pdfPath.toFile()));

    System.out.println("====== 📄 PDF Loader Success ======");
    
    // पूर्ण PDF चा टेक्स्ट मिळवणे
    String allContent = document.text();
    
    System.out.println("Total text length loaded: " + allContent.length() + " characters.");
    System.out.println("\n--- First page preview (first 500 chars) ---");

    // ४. पहिले ५०० कॅरेक्टर्स सबस्ट्रिंग वापरून प्रिंट करणे [ :500 ]
    if (allContent.length() >= 500) {
        System.out.println(allContent.substring(0, 500));
    } else {
        System.out.println(allContent);
    }

} catch (Exception e) {
    System.out.println("❌ PDF लोड करताना एरर आली: " + e.getMessage());
    e.printStackTrace();
}
```

```java
import java.io.BufferedReader;
import java.io.InputStreamReader;

try {
    // १. तुझा अचूक पीडीएफ पाथ
    String pdfPath = "E:\\Agentic AI\\4_rag_basic\\telecom_guide.pdf";

    // २. जावा मधूनच बॅकग्राउंडला पायथनची स्क्रिप्ट चालवणे (जी पीडीएफ रीड करेल)
    String pythonCode = "import pypdf; "
                      + "reader = pypdf.PdfReader(r'" + pdfPath + "'); "
                      + "pages_count = len(reader.pages); "
                      + "print(f'Loaded {pages_count} pages from the PDF.'); "
                      + "print('\\n--- First page preview (first 500 chars) ---'); "
                      + "print(reader.pages[0].extract_text()[:500])";
    
    ProcessBuilder pb = new ProcessBuilder("python", "-c", pythonCode);
    Process process = pb.start();

    // ३. पायथनने दिलेला आउटपुट जावाने स्क्रीनवर प्रिंट करणे
    BufferedReader reader = new BufferedReader(new InputStreamReader(process.getInputStream()));
    String line;
    System.out.println("====== 📄 PDF Loader Success (Java-Python Bypass) ======");
    while ((line = reader.readLine()) != null) {
        System.out.println(line);
    }

    // जर काही एरर आली तर ती दाखवण्यासाठी
    BufferedReader errReader = new BufferedReader(new InputStreamReader(process.getErrorStream()));
    while ((line = errReader.readLine()) != null) {
        System.out.println("⚠️ Notice: " + line);
    }

    process.waitFor();

} catch (Exception e) {
    System.out.println("❌ एरर आली: " + e.getMessage());
}
```

```java
import java.io.BufferedReader;
import java.io.InputStreamReader;
import java.io.OutputStream;
import java.net.HttpURLConnection;
import java.net.URL;
import java.nio.charset.StandardCharsets;

try {
    // १. तुझा अचूक पीडीएफ पाथ आणि क्रोमाचा बेस URL
    String pdfPath = "E:\\Agentic AI\\4_rag_basic\\telecom_guide.pdf";
    String chromaBaseUrl = "http://127.0.0.1:8000/api/v1";

    // २. पायथनच्या मदतीने पीडीएफ पूर्ण रीड करणे
    String pythonCode = "import pypdf; "
                      + "reader = pypdf.PdfReader(r'" + pdfPath + "'); "
                      + "text = ''.join([page.extract_text() for page in reader.pages]); "
                      + "print(text)";
    
    ProcessBuilder pb = new ProcessBuilder("python", "-c", pythonCode);
    Process process = pb.start();

    BufferedReader pdfReader = new BufferedReader(new InputStreamReader(process.getInputStream(), StandardCharsets.UTF_8));
    StringBuilder fullPdfText = new StringBuilder();
    String line;
    while ((line = pdfReader.readLine()) != null) {
        fullPdfText.append(line).append(" ");
    }
    process.waitFor();

    String extractedText = fullPdfText.toString().trim();
    System.out.println("====== 📄 PDF Loading Completed ======");
    System.out.println("Extracted Length: " + extractedText.length() + " characters.");

    if (extractedText.isEmpty()) {
        System.out.println("⚠️ पीडीएफ मधून डेटा मिळाला नाही. कृपया पाथ तपासा.");
    } else {
        // ३. 'telecom_collection' नावाचे नवीन कलेक्शन क्रोमामध्ये तयार करणे
        URL createUrl = new URL(chromaBaseUrl + "/collections");
        HttpURLConnection connCreate = (HttpURLConnection) createUrl.openConnection();
        connCreate.setRequestMethod("POST");
        connCreate.setRequestProperty("Content-Type", "application/json");
        connCreate.setDoOutput(true);

        String createJson = "{\"name\": \"telecom_collection\", \"get_or_create\": true}";
        try (OutputStream os = connCreate.getOutputStream()) {
            byte[] input = createJson.getBytes(StandardCharsets.UTF_8);
            os.write(input, 0, input.length);
        }

        BufferedReader brCreate = new BufferedReader(new InputStreamReader(connCreate.getInputStream(), StandardCharsets.UTF_8));
        StringBuilder respCreate = new StringBuilder();
        while ((line = brCreate.readLine()) != null) { respCreate.append(line.trim()); }
        brCreate.close();

        String collectionId = respCreate.toString().split("\"id\"\\s*:\\s*\"")[1].split("\"")[0];
        System.out.println("✅ Chroma Collection Ready. ID: " + collectionId);
        
        // ४. डेटा चंकींग आणि इन्सर्शनचा पुढचा भाग आपण आता करू शकतो!
    }

} catch (Exception e) {
    System.out.println("❌ एरर आली: " + e.getMessage());
    e.printStackTrace();
}
```
