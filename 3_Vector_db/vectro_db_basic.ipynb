```java
import java.io.BufferedReader;
import java.io.InputStreamReader;
import java.io.OutputStream;
import java.net.HttpURLConnection;
import java.net.URL;
import java.nio.charset.StandardCharsets;
import java.util.regex.Matcher;
import java.util.regex.Pattern;

try {
    String baseUrl = "http://127.0.0.1:8000/api/v1";

    // Step 1: आधी 'news_articles' चा Collection ID शोधणे
    URL createUrl = new URL(baseUrl + "/collections");
    HttpURLConnection conn1 = (HttpURLConnection) createUrl.openConnection();
    conn1.setRequestMethod("POST");
    conn1.setRequestProperty("Content-Type", "application/json");
    conn1.setDoOutput(true);

    String createJson = "{\"name\": \"news_articles\", \"get_or_create\": true}";
    try (OutputStream os = conn1.getOutputStream()) {
        byte[] input = createJson.getBytes(StandardCharsets.UTF_8);
        os.write(input, 0, input.length);
    }

    StringBuilder response1 = new StringBuilder();
    try (BufferedReader br = new BufferedReader(new InputStreamReader(conn1.getInputStream(), StandardCharsets.UTF_8))) {
        String responseLine;
        while ((responseLine = br.readLine()) != null) {
            response1.append(responseLine.trim());
        }
    }
    
    String collectionId = response1.toString().split("\"id\"\\s*:\\s*\"")[1].split("\"")[0];

    // Step 2: कलेक्शनमधून सर्व डेटा आणि एम्बेडिंग्स फेच करणे (पायथनच्या collection.get सारखं)
    URL getUrl = new URL(baseUrl + "/collections/" + collectionId + "/get");
    HttpURLConnection conn2 = (HttpURLConnection) getUrl.openConnection();
    conn2.setRequestMethod("POST"); // ChromaDB चा GET डेटा पोस्ट रिक्वेस्टने मिळतो
    conn2.setRequestProperty("Content-Type", "application/json");
    conn2.setDoOutput(true);

    // इमेजमधील include=["documents", "embeddings", "metadatas"] सारखाच JSON पेलोड
    String getJson = "{\"include\": [\"documents\", \"embeddings\", \"metadatas\"]}";
    try (OutputStream os = conn2.getOutputStream()) {
        byte[] input = getJson.getBytes(StandardCharsets.UTF_8);
        os.write(input, 0, input.length);
    }

    StringBuilder response2 = new StringBuilder();
    try (BufferedReader br = new BufferedReader(new InputStreamReader(conn2.getInputStream(), StandardCharsets.UTF_8))) {
        String responseLine;
        while ((responseLine = br.readLine()) != null) {
            response2.append(responseLine.trim());
        }
    }
    String jsonResponse = response2.toString();

    System.out.println("====== Fetching Documents & Embeddings ======");

    // Step 3: रेग्युलर एक्सप्रेशन (Regex) वापरून डॉक्युमेंट्स आणि एम्बेडिंग्स वेगळे करणे
    Pattern docPattern = Pattern.compile("\"documents\"\\s*:\\s*\\[(.*?)\\]");
    Pattern embPattern = Pattern.compile("\"embeddings\"\\s*:\\s*\\[\\[(.*?)\\]\\]");

    Matcher docMatcher = docPattern.matcher(jsonResponse);
    Matcher embMatcher = embPattern.matcher(jsonResponse);

    if (docMatcher.find() && embMatcher.find()) {
        // सर्व डॉक्युमेंट्स स्प्लिट करणे
        String[] docs = docMatcher.group(1).split("\",\\s*\"");
        // सर्व एम्बेडिंग्स स्प्लिट करणे
        String[] embs = embMatcher.group(1).split("\\],\\s*\\[");

        for (int i = 0; i < docs.length; i++) {
            String text = docs[i].replace("\"", "");
            System.out.println("\nDocument " + (i + 1));
            System.out.println("Text: " + text);

            if (i < embs.length) {
                String[] vectorValues = embs[i].split(",");
                // 1. Embedding Length प्रिंट करणे
                System.out.println("Embedding length: " + vectorValues.length);

                // 2. पहिले 10 व्हॅल्यूज प्रिंट करणे (पायथनच्या emb[:10] सारखं)
                System.out.print("First 10 values: [");
                for (int j = 0; j < Math.min(10, vectorValues.length); j++) {
                    System.out.print(vectorValues[j].trim() + (j < 9 ? ", " : ""));
                }
                System.out.println("...]");
            }
        }
    } else {
        System.out.println("⚠️ कलेक्शन रिकामे आहे किंवा डेटा सापडला नाही. आधीचा डेटा ॲड करण्याचा कोड पुन्हा रन करा.");
    }

} catch (Exception e) {
    System.out.println("❌ डेटा फेच करताना एरर आली: " + e.getMessage());
    e.printStackTrace();
}
```

```java
import dev.langchain4j.data.segment.TextSegment;
import dev.langchain4j.model.embedding.EmbeddingModel;
import dev.langchain4j.model.embedding.onnx.allminilml6v2.AllMiniLmL6V2EmbeddingModel;
import java.io.BufferedReader;
import java.io.InputStreamReader;
import java.io.OutputStream;
import java.net.HttpURLConnection;
import java.net.URL;
import java.nio.charset.StandardCharsets;
import java.util.regex.Matcher;
import java.util.regex.Pattern;

try {
    String baseUrl = "http://127.0.0.1:8000/api/v1";
    String collectionId = "e2bc480d-dd7f-453b-9927-80c34d7e642f"; // तुमचा चालू कलेक्शन आयडी

    // 1. एम्बेडिंग मॉडेल वापरून क्वेरी टेक्स्टचे व्हेक्टर तयार करा
    EmbeddingModel embeddingModel = new AllMiniLmL6V2EmbeddingModel();
    String queryText = "I Love IPad";
    float[] vector = embeddingModel.embed(TextSegment.from(queryText)).content().vector();

    // 2. क्रोमाच्या गरजेनुसार [ [vector_values] ] असा JSON नेस्टेड ॲरे बनवणे
    StringBuilder queryEmbeddingJson = new StringBuilder("[[");
    for (int j = 0; j < vector.length; j++) {
        queryEmbeddingJson.append(vector[j]).append(j < vector.length - 1 ? "," : "");
    }
    queryEmbeddingJson.append("]]");

    // 3. थेट क्रोमाच्या /query एंडपॉईंटला कनेक्ट करा
    URL queryUrl = new URL(baseUrl + "/collections/" + collectionId + "/query");
    HttpURLConnection conn = (HttpURLConnection) queryUrl.openConnection();
    conn.setRequestMethod("POST");
    conn.setRequestProperty("Content-Type", "application/json");
    conn.setDoOutput(true);

    // पायथनच्या n_results=2 सारखाच परफेक्ट पेलोड (फक्त documents आणि distances मागवले)
    String queryJson = "{"
            + "\"query_embeddings\": " + queryEmbeddingJson.toString() + ","
            + "\"n_results\": 2,"
            + "\"include\": [\"documents\", \"distances\"]"
            + "}";

    try (OutputStream os = conn.getOutputStream()) {
        byte[] input = queryJson.getBytes(StandardCharsets.UTF_8);
        os.write(input, 0, input.length);
    }

    // 4. रिस्पॉन्स वाचणे
    BufferedReader br = new BufferedReader(new InputStreamReader(conn.getInputStream(), StandardCharsets.UTF_8));
    StringBuilder response = new StringBuilder();
    String line;
    while ((line = br.readLine()) != null) {
        response.append(line.trim());
    }
    br.close();
    String jsonResponse = response.toString();

    System.out.println("====== 🔍 Semantic Search Results ======");

    // Regex वापरून मॅच झालेले डॉक्युमेंट्स आणि डिस्टन्स वेगळे करणे
    Pattern docPattern = Pattern.compile("\"documents\"\\s*:\\s*\\[\\[(.*?)\\]\\]");
    Pattern distPattern = Pattern.compile("\"distances\"\\s*:\\s*\\[\\[(.*?)\\]\\]");

    Matcher docMatcher = docPattern.matcher(jsonResponse);
    Matcher distMatcher = distPattern.matcher(jsonResponse);

    if (docMatcher.find()) {
        String[] docs = docMatcher.group(1).split("\",\\s*\"");
        String[] distances = distMatcher.find() ? distMatcher.group(1).split(",") : new String[docs.length];

        for (int i = 0; i < docs.length; i++) {
            String text = docs[i].replace("\"", "");
            System.out.println("\nMatch " + (i + 1));
            System.out.println("Text: " + text);
            if (i < distances.length) {
                System.out.println("Distance Score: " + distances[i].trim());
            }
        }
    } else {
        System.out.println("⚠️ कोणतेही मिळतेजुळते डॉक्युमेंट्स सापडले नाहीत.");
    }

} catch (Exception e) {
    System.out.println("❌ एरर आली: " + e.getMessage());
    e.printStackTrace();
}
```
