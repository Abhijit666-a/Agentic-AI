```java
import dev.langchain4j.model.openai.OpenAiChatModel;
import dev.langchain4j.model.chat.ChatLanguageModel;
import java.io.*;
import java.nio.file.*;
```

```java
String apiKey = "AIzaSyDNsoTkxbTCpdpFMUsfcrFu3uo-YBVTc_M";

ChatLanguageModel llm = OpenAiChatModel.builder()
    .apiKey(apiKey)
    .baseUrl("https://generativelanguage.googleapis.com/v1beta/openai/")
    .modelName("gemini-2.5-flash")
    .build();

String bloodReport = Files.readString(Paths.get("blood_work.txt"));

String extractionPrompt = """
    You are a medical data extraction assistant.

    From the blood report below, extract ALL test values and classify each one as HIGH, LOW, or NORMAL
    based on the reference ranges provided in the report.

    Format your response as:
    - Test Name: value | Status: HIGH/LOW/NORMAL | Reference: range

    Blood Report:
    """ + bloodReport;

String extractedValues = llm.generate(extractionPrompt);

System.out.println("=== STAGE 1: EXTRACTED VALUES ===");
System.out.println(extractedValues);
```
